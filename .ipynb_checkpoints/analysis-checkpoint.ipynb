```python
import pandas as pd
import folium
```

```python
# Adjust path if your CSV is in a "data" folder
df = pd.read_csv("data/MTA_Bus_Automated_Camera_Enforcement_Violations__Beginning_October_2019_20250919.csv")

# Quick check
df.head()
```

```python
# Only keep rows where Violation Status contains "EXEMPT"
exempt_df = df[df["Violation Status"].str.contains("EXEMPT", na=False)]

# Quick check
exempt_df.head()
```

```python
# Count how many times each Vehicle ID appears in exempt violations
repeat_offenders = (
    exempt_df.groupby("Vehicle ID")
    .size()
    .reset_index(name="violation_count")
    .sort_values(by="violation_count", ascending=False)
)

# Show top 10 repeat offenders
repeat_offenders.head(10)
```

```python
# Take only top 10 repeat offenders
top_offenders = repeat_offenders.head(10)

# Filter original exempt_df for these Vehicle IDs
top_exempt_df = exempt_df[exempt_df["Vehicle ID"].isin(top_offenders["Vehicle ID"])]

top_exempt_df.head()
```

```python
# NYC-centered map
m = folium.Map(location=[40.75, -73.97], zoom_start=11)

for _, row in top_exempt_df.iterrows():
    if pd.notna(row["Violation Latitude"]) and pd.notna(row["Violation Longitude"]):
        folium.CircleMarker(
            location=[row["Violation Latitude"], row["Violation Longitude"]],
            radius=3,
            color="red",
            fill=True,
            fill_opacity=0.6,
        ).add_to(m)

# Save to HTML to view
m.save("top_exempt_violations_map.html")
print("✅ Map saved! Open top_exempt_violations_map.html in your browser to view it.")
```
